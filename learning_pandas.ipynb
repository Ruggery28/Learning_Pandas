```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.info() #getting all the information about the data
```

```python
df.head() #method head shows the first 5 rows of the table (how much you want)
```

```python
#We can use isnull and combine with sum to see how many rows are empty 
df.isnull().sum()           #it will show all collumns
df['Age'].isnull().sum()    #it will show only 'Age'
df.isnull().sum().sort_values(ascending=False) #can sort it False from the high to low
```

```python
#To see the values that is missing
missing_values = df[df['Age'].isnull()]
missing_values.head()
```

```python
#show only the collumns you want
df[['Age', 'Name']].head()
```

```python
#FILTERING WITH CONDITIONS
df[df['Age'] > 30].head(3) #filtering with one condition
df[(df['Age'] > 30) & (df['Sex'] == 'male')].head() #filtering with more than one conditions
```

```python
#Filtering text
df[df['Name'].str.contains('Mrs')].head(3)
```

```python
#Droping rowns where is missing values
df.dropna(subset=['Age']).head()
```

```python
#Filling missing values - Replace missing values with a number, string, or calculation:
df['Age'] = df['Age'].fillna(df['Age'].median())  # median age
df['Cabin'] = df['Cabin'].fillna('Unknown')       # label missing cabins
df.head()
```

```python
#Combination of all I have learned: Filling missing values, Renaming columns, changing types, sorting vales and groupby
df = (
    df.fillna({'Age': df['Age'].median(), 'Cabin': 'Unknown', 'Embarked': 'Unknown'})
    .rename(columns={'Pclass': 'PassengerClass', 'Fare': 'TicketFare'})
    .astype({'TicketFare': float})
    .sort_values(by=['PassengerClass', 'TicketFare'], ascending=[True, False])
    .groupby('PassengerClass').head(3) # keeps only top rows without grouping index
    [['PassengerClass', 'Name', 'Age', 'TicketFare', 'Cabin']]
    
)
df
```
